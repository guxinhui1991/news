```python
import numpy as np
import pandas as pd
import statsmodels as sm
import scipy.stats as stats
import os
import urllib
import csv
from bs4 import BeautifulSoup
import codecs

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Headlines.csv")
data=df.as_matrix()
data=np.asmatrix(data)
data=data[:,1:3]
good_words=['good','growth','How']
bad_words=['bad','decline','never']
temp=data[0,1]
k=0
good_number=[]
bad_number=[]
for i in range(data.shape[0]):
    if data[i,1]==temp: continue
    if data[i,1]!=temp:
        X=data[k:i,0]
        k=i+1
        temp=data[k,1]
        good=0
        bad=0
        for m in range(X.shape[1]):
            string=X[m,0]
            for j in range(len(good_words)):
                if good_words[j] in string:
                    good=good+1
                if bad_words[j] in string:
                    bad=bad+1
        good_number.append(good)
        bad_number.append(bad)
total=[]
for i in range(len(good_number)):
    total.append(good_number[i]-bad_number[i])
#good_number and bad_number are a list of number of good/bad words apperance each day, 
#take the difference and form total, for day i, if total[i]>0 means there are more good
#words than bad occured on that day and vise versa. As a baseline we can just add like 
#20 words to the dictionary. I put 'how' in just as a test because i know that's in the 
#first news.
```
